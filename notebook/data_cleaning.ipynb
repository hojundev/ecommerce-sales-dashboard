```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', None)
print("Setup successful")
```

```python
try:
    df = pd.read_csv('../data/online_retail_II.csv')
except UnicodeDecodeError:
    df = pd.read_csv('../data/online_reatil_II.csv', encoding='ISO-8859-1')

print(f"Dimensions: {df.shape}")
df.head()
```

```python
print("--- Null Value Count ---")
print(df.isnull().sum())

print("\n--- Duplicate Count ---")
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
# cleaning data
df_clean = df.dropna(subset=['Customer ID']).copy()
df_clean.drop_duplicates(inplace=True)
df_sales = df_clean[df_clean['Quantity'] > 0].copy()
print(f"Rows after cleaning: {len(df_sales)}")
```

```python
country_map = {
    'EIRE': 'Ireland',
    'RSA': 'South Africa'
}

df_sales['Country'] = df_sales['Country'].replace(country_map)
print("Unique Countries (Sample): ", df_sales['Country'].unique()[:10])
```

```python
df_sales['InvoiceDate'] = pd.to_datetime(df_sales['InvoiceDate'])
df_sales['TotalAmount'] = df_sales['Quantity'] * df_sales['Price']
df_sales['Customer ID'] = df_sales['Customer ID'].astype(int)
df_sales.head()
```

```python
df_sales.to_csv('../data/cleaned_online_retail.csv', index=False)
print("File saved")
```
